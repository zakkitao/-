```python
# 实例3 百度、360搜索关键词提交
## 百度的搜索关键词接口是：http://www.baidu.com/s？wd=keyword
## 360的抖索关键词接口是： http://www.so.com/s?q=keyword
import requests
kv = {"q":"陶伊丽"}
def getwd():
    try:
        r = requests.get(url,params = kv)
        r.raise_for_status()
        r.encoding = r.apparent_encoding
        return r.text[:1000]
    except:
        return "发生错误！"
if __name__ == "__main__":
    url = "http://www.so.com/s"
    print(getwd())
```
